### About the data
- The dataset contains informationa about admissions in the university based on several parameters, which includes:
- Serial No : To uniquely indentify students
- GRE Score : Score of GRE test which is an important test for admissions in the graduate school or business school application process globally.
- TOEFL Score : Test of English as a Foreign Language exam score
- University Rating : Rating of the University out of 5
- SOP : Related to Statement of Purpose(SOP) for applying to a particular course or university.
- LOR : Some score related to LOR i.e a letter of recommendation
- CGPA : CGPA is a past performance measure of aspirant
- Research : Binary values of either 1 or 0
- Chance of Admit : Probability of the student to get admission the university

### Analysing the Data

#### Heatmap showing the correlation among the parameters for admission

```python
sns.heatmap(data= adm.corr(), annot=True, linecolor='white')
plt.title('Correlation of parameters for admission', fontweight='bold', fontsize=16)
```

#### Observation
- The varying intensity of color represents the measure of correlation. Correlation is a measure of linear relationship between two variables.Correlation ranges between 1 and -1. Values closer to 0 means there is no linear trend between the two variables. The close to 1 the correlation is the more positively correlated they are; that is as one increases so does the other and the closer to 1 the stronger this relationship is. A correlation closer to -1 is similar, but instead of both increasing, one variable will decrease as the other increases.
- It is seen that 'Research' is least correlated with other parameters
- The correlation between GRE score and TOEFL is 0.84, this shows a strong relationship

### Regression plot (lmplot) for showing correlation between Chance of admission on other parameters

```python
sns.regplot(x='GRE Score',y='Chance of Admit ',data=adm,scatter_kws={'color':'purple'}, line_kws={'color':'purple'})
plt.title('Chance of Admit vs GRE Score',fontweight='bold', fontsize= 16)
```

#### Observation
- The relationship between GRE score and chance of Admit is a positive linear relationship
- The student has a higher chance of being admitted as the GRE score increases
- The chance of being admitted is the dependent-variable and the GRE score is the independent variable

```python
sns.lmplot(x='GRE Score',y='Chance of Admit ',data=adm, hue='Research')
plt.title('Chance of Admit vs GRE Score Taking Research into Consideration',fontweight='bold', fontsize= 16)
```

#### Observation
- This shows that students who carried out research had higher scores in GRE, giving them a higer chance of being admitted

```python
sns.lmplot(x='TOEFL Score',y='Chance of Admit ',data=adm, hue='Research')
plt.title('Chance of Admit vs TOEFL Score Taking Research into Consideration',fontweight='bold', fontsize= 16)
```

#### Observation
- The relationship between TOEFL score and chance of Admit is a positive linear relationship
- As the TOEFL score increases, students have a higher chance of being admitted
- A higher number of students who had high TOEFL also carried out research and had higher chance of being admitted

```python
sns.lmplot(x='CGPA',y='Chance of Admit ',data=adm, hue= 'Research')
plt.title('Chance of Admit vs CGPA Taking Research into Consideration',fontweight='bold', fontsize= 16)
```

#### Observation
- The relationship between CGPA and chance of Admit is a positive linear relationship
- As the CGPA score increases, the student has a higher chance of being admitted
- Also,many students who had high CGPA carried out research and had a higher chance of being admitted 

```python
sns.lmplot(x='SOP',y='Chance of Admit ',data=adm, hue='Research')
plt.title('Chance of Admit vs SOP Taking Research into Consideration',fontweight='bold', fontsize= 16)
```

#### Observation
- This shows the relationship between SOP score and chance of Admit
- As the SOP score increases, the student has a higher chance of being admitted. 
- However, it is seen here that for some students, despite having a high SOP score, they still had a lesser chance of being admitted.
- Students with high SOP scores carried out research and had a high chance of being admitted.

```python
sns.regplot(x='Research',y='Chance of Admit ',data=adm, 
            scatter_kws={'color':'teal','alpha':0.6},line_kws={'color':'teal','alpha':0.6})
plt.title('Chance of Admit vs Research',fontweight='bold', fontsize= 16)
```

#### Observation
- This shows the relationship between Research and chance of Admit
- Majority of students with research had higher chances of being admitted than students with no research.
- However, some students who had research still had a low chance of being admitted and few with no research also stood a moderate chance of gaining admission
